Perhitungan turunan parsial/vector gradient

```python
R = picture[:,:,0]
R_dx,R_dy = np.gradient(R)

fig, ax = plt.subplots(ncols=3, figsize=(21,5))
ax[0].imshow(R)
ax[0].set_title('Gambar Filter Red')
ax[1].imshow(R_dx)
ax[1].set_title('dx')
ax[2].imshow(R_dy)
ax[2].set_title('dy')
```

```python
matriks_random=np.random.rand(780,300)
```

Perkalian matriks

```python
matriks_random
```

Visualisasi data

```python
magnitude = np.sqrt ((R_dx**2) + (R_dy**2))
plt.figure(figsize = (21,5))
plt.imshow(magnitude, cmap='gray')
print('max',magnitude.max())
print('min', magnitude.min())
```

```python
hasil_perkalian_matriks = magnitude.dot(matriks_random)
print(hasil_perkalian_matriks)
```
