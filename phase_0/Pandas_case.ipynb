Grouping and Aggregation

```python
df.groupby('fran_id')['pts'].sum()
```

```python
df.groupby('fran_id')['pts'].mean()
```

```python
df[
    (df['fran_id'] == 'Spurs') &
    (df['year_id'] > 2010)
].groupby('year_id')['pts'].sum()
```

```python
df[
    (df['fran_id'] == 'Spurs') &
    (df['year_id'] > 2010)
].groupby(['year_id', 'game_result' ])['pts'].sum()
```

```python
df[
    (df['fran_id'] == 'Spurs') &
    (df['year_id'] > 2010)
].groupby(['game_result', 'year_id' ])['pts'].sum()
```
